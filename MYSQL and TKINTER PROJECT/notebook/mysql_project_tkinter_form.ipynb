# This code creates a simple GUI with three tabs: Books, Customers, and Orders. Each tab has fields for entering data and a button to add the data to the corresponding table in the database.

```python
from tkinter import *
from tkinter import messagebox
import pymysql
```

```python
connection = pymysql.connect(
    host = "localhost",
    user = "root",
    password = "@123454",
    db = "MySQL_Project"
)


cursor = connection.cursor()

# Create Tkinter window
root = tk.Tk()
root.title("Bookstore Management System")

# Notebook for tabs
notebook = ttk.Notebook(root)
notebook.pack(pady=10, expand=True)

# Frames for tabs
frame1 = tk.Frame(notebook)
frame2 = tk.Frame(notebook)
frame3 = tk.Frame(notebook)

# Add frames to notebook
notebook.add(frame1, text="Books")
notebook.add(frame2, text="Customers")
notebook.add(frame3, text="Orders")

# Books tab
tk.Label(frame1, text="Book ID").grid(row=0, column=0)
tk.Label(frame1, text="Title").grid(row=1, column=0)
tk.Label(frame1, text="Author").grid(row=2, column=0)
tk.Label(frame1, text="Genre").grid(row=3, column=0)
tk.Label(frame1, text="Published Year").grid(row=4, column=0)
tk.Label(frame1, text="Price").grid(row=5, column=0)
tk.Label(frame1, text="Stock").grid(row=6, column=0)

book_id = tk.Entry(frame1)
title = tk.Entry(frame1)
author = tk.Entry(frame1)
genre = tk.Entry(frame1)
published_year = tk.Entry(frame1)
price = tk.Entry(frame1)
stock = tk.Entry(frame1)

book_id.grid(row=0, column=1)
title.grid(row=1, column=1)
author.grid(row=2, column=1)
genre.grid(row=3, column=1)
published_year.grid(row=4, column=1)
price.grid(row=5, column=1)
stock.grid(row=6, column=1)

def add_book():
    query = "INSERT INTO Books (Book_ID, Title, Author, Genre, Published_Year, Price, Stock) VALUES (%s, %s, %s, %s, %s, %s, %s)"
    values = (book_id.get(), title.get(), author.get(), genre.get(), published_year.get(), price.get(), stock.get())
    cursor.execute(query, values)
    connection.commit()
    messagebox.showinfo("Success", "Book added successfully")

tk.Button(frame1, text="Add Book", command=add_book).grid(row=7, column=0, columnspan=2)

# Customers tab
tk.Label(frame2, text="Customer ID").grid(row=0, column=0)
tk.Label(frame2, text="Name").grid(row=1, column=0)
tk.Label(frame2, text="Email").grid(row=2, column=0)
tk.Label(frame2, text="Phone").grid(row=3, column=0)
tk.Label(frame2, text="City").grid(row=4, column=0)
tk.Label(frame2, text="Country").grid(row=5, column=0)

customer_id = tk.Entry(frame2)
name = tk.Entry(frame2)
email = tk.Entry(frame2)
phone = tk.Entry(frame2)
city = tk.Entry(frame2)
country = tk.Entry(frame2)

customer_id.grid(row=0, column=1)
name.grid(row=1, column=1)
email.grid(row=2, column=1)
phone.grid(row=3, column=1)
city.grid(row=4, column=1)
country.grid(row=5, column=1)

def add_customer():
    query = "INSERT INTO Customers (Customer_ID, Name, Email, Phone, City, Country) VALUES (%s, %s, %s, %s, %s, %s)"
    values = (customer_id.get(), name.get(), email.get(), phone.get(), city.get(), country.get())
    cursor.execute(query, values)
    connection.commit()
    messagebox.showinfo("Success", "Customer added successfully")

tk.Button(frame2, text="Add Customer", command=add_customer).grid(row=6, column=0, columnspan=2)

# Orders tab
tk.Label(frame3, text="Order ID").grid(row=0, column=0)
tk.Label(frame3, text="Customer ID").grid(row=1, column=0)
tk.Label(frame3, text="Book ID").grid(row=2, column=0)
tk.Label(frame3, text="Order Date").grid(row=3, column=0)
tk.Label(frame3, text="Quantity").grid(row=4, column=0)
tk.Label(frame3, text="Total Amount").grid(row=5, column=0)

order_id = tk.Entry(frame3)
customer_id_order = tk.Entry(frame3)
book_id_order = tk.Entry(frame3)
order_date = tk.Entry(frame3)
quantity = tk.Entry(frame3)
total_amount = tk.Entry(frame3)

order_id.grid(row=0, column=1)
customer_id_order.grid(row=1, column=1)
book_id_order.grid(row=2, column=1)
order_date.grid(row=3, column=1)
quantity.grid(row=4, column=1)
total_amount.grid(row=5, column=1)


def add_order():
    query = "INSERT INTO Orders (Order_ID, Customer_ID, Book_ID, Order_Date, Quantity, Total_Amount) VALUES (%s, %s, %s, %s, %s, %s)"
    values = (order_id.get(), customer_id_order.get(), book_id_order.get(), order_date.get(), quantity.get(), total_amount.get())
    cursor.execute(query, values)
    connection.commit()
    messagebox.showinfo("Success", "Order added successfully")

tk.Button(frame3, text="Add Order", command=add_order).grid(row=6, column=0, columnspan=2)

root.mainloop()
```
